#### Find the mean and standard deviation of the temperature of each state.

```python
connecticut = temp['State'] == 'Connecticut'
mean_connecticut = str(temp.loc[connecticut, 'Temperature'].mean())
std_connecticut = str(temp.loc[connecticut, 'Temperature'].std())
print("Connecticut mean temperature is: " + mean_connecticut)
print("Connecticut std deviation of the temperature is: " + std_connecticut)

ny = temp['State'] == 'New York'
mean_ny = str(temp.loc[ny, 'Temperature'].mean())
std_ny = str(temp.loc[ny, 'Temperature'].std())
print("New York mean temperature is: " + mean_ny)
print("New York std deviation of the temperature is: " + std_ny)

new_jersey = temp['State'] == 'New Jersey'
mean_jersey = str(temp.loc[new_jersey, 'Temperature'].mean())
std_jersey = str(temp.loc[new_jersey, 'Temperature'].std())
print("Connecticut mean temperature is: " + mean_jersey)
print("Connecticut std deviation of the temperature is: " + std_jersey)
```
